```python
import pandas as pd
import numpy as np
import folium
```

```python
import json

with open('bairros2.json', 'r', encoding='utf-8-sig') as f:
   maprecife = json.load(f)
```

```python
#Latitude: -8.05428, Longitude: -34.8813
recife = folium.Map(location=[-8.05428, -34.8813],tiles='OpenStreetMap', zoom_start=12)
folium.GeoJson(
    maprecife
).add_to(recife)
recife
#recife.save('index.html')
```

```python
bairros = pd.read_csv('http://dados.recife.pe.gov.br/dataset/2a9b1c39-0700-4ddf-9a10-b3c8d5d9396c/resource/b686ebbf-5f54-4116-9227-0be464b9aedd/download/casos-dengue2015.csv',
                sep=';').dropna(how='all', axis = 1)
bairros.head()
```

```python
bairrin = pd.DataFrame()
bairrin['co_bairro_residencia'] = bairros['co_bairro_residencia']
bairrin['no_bairro_residencia'] = bairros['no_bairro_residencia']
bairrin=bairrin.drop_duplicates()
bairrin.sort_values(by='co_bairro_residencia')
bairrin = bairrin.drop(6379)
bairrin['co_bairro_residencia'] = bairrin['co_bairro_residencia'] * 100
bairrin
```

```python
m = folium.Map(location=[-8.05428, -34.88132], tiles="OpenStreetMap", zoom_start=13)
m.choropleth(geo_data=maprecife, data=bairrin,
             columns=['no_bairro_residencia', 'co_bairro_residencia'],
             key_on='feature.bairro_codigo',
             fill_color='YlGn', fill_opacity=0.6, line_opacity=0.2,
             legend_name='Bairro Nomes')
m
```
