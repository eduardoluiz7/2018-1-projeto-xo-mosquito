# Projeto xo-mosquito

## Proposta
Levando em consideração o contexto das doenças tropicais negligenciadas, em
especial Dengue, Zika e Chikungunya, e seu impacto social e, segundo IANPHI (2016),
econômico na américa latina, especialmente no Brasil, o projeto possui como proposta:
- Analisar a relação entre o perfil social dos indivíduos afetados e as incidências de doenças transmitidas pelo mosquito Aedes Aegypti;
- Investigar a influência da localidade física e climáticos em que se encontram tais indivíduos;
- Encontrar padrões nos casos relatados para assim sugerir abordagens de vigilância que possam ser implementados por governos.

O objetivo é analisar essas hipóteses e tentar validá-las ou não em relação a Recife,
Pernambuco, mas se não for possível, devido a falta de dados, então o escopo será analisar
globalmente.

```python
def raca(tipo_raca):
    if(tipo_raca == 1):
        return 'branca'
    elif(tipo_raca == 2):
        return 'preta'
    elif(tipo_raca == 3):
        return 'amarela'
    elif(tipo_raca == 4):
        return 'parda'
    elif(tipo_raca == 5):
        return 'indigena'
    elif(tipo_raca == 9):
        return 'ignorado'

df_estudo['tp_raca_cor'] = df_estudo.apply(lambda x: raca(x['tp_raca_cor']), axis = 1)
# '1- branca, 2- preta, 3- '
#                                                  'amarela, 4- parda, 5- '
#                                                  'indígena, 9 Ignorado'
```

```python
def escolaridade(nivel):
    if(nivel == 43):
        return 'Analfabeto'
    elif(nivel == 1):
        return '1 a 4 serie'
    elif(nivel == 2):
        return '4 serie completa'
    elif(nivel == 3):
        return '5 a 8 serie'
    elif(nivel == 4):
        return '8 serie completa'
    elif(nivel == 5):
        return 'ensino medio incompleto'
    elif(nivel == 6):
        return 'ensino medio completo'
    elif(nivel == 7):
        return 'ensino superior incompleto'
    elif(nivel == 8):
        return 'ensino superior completo'
    elif(nivel == 9):
        return 'ignorado'
    elif(nivel == 10):
        return 'nao se aplica'


df_estudo['tp_escolaridade'] = df_estudo.apply(lambda x: escolaridade(x['tp_escolaridade']), axis = 1)

# '43 Analfabeto, 1- 1ª a 4ª '
#                                                  'série incompleta do EF, 2- '
#                                                  '4ª série completa do EF '
#                                                  '(antigo 1° grau), 3- 5ª à 8ª '
#                                                  'série incompleta do EF '
#                                                  '(antigo ginásio ou 1°grau), '
#                                                  '4- Ensino fundamental '
#                                                  'completo (antigo ginásio ou '
#                                                  '1° grau), 5- Ensino médio '
#                                                  'incompleto(antigo colegial '
#                                                  'ou 2° grau), 6- Ensino médio '
#                                                  'completo (antigo colegial ou '
#                                                  '2° grau), 7- Educação '
#                                                  'superior incompleta, 8- '
#                                                  'Educação superior completa, '
#                                                  '9- Ignorado, 10- Não se '
#                                                  'aplica'},
```
